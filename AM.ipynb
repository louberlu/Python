```python
import numpy as np
import matplotlib.pyplot as plt

# Paramètres du signal
fc = 10  # fréquence de la porteuse
fm = 2   # fréquence du signal modulant
m = 0.5  # indice de modulation

# Génération du temps
t = np.linspace(0, 2, 1000)

# Génération des signaux
porteuse = np.cos(2 * np.pi * fc * t)
modulant = m * np.cos(2 * np.pi * fm * t)
signal_AM = (1 + modulant) * porteuse

# Affichage des signaux
plt.figure(figsize=(10, 6))
plt.subplot(3, 1, 1)
plt.plot(t, porteuse)
plt.title('Signal de la porteuse')
plt.subplot(3, 1, 2)
plt.plot(t, modulant)
plt.title('Signal modulant')
plt.subplot(3, 1, 3)
plt.plot(t, signal_AM)
plt.title('Signal modulé (résultat de la modulation d\'amplitude)')
plt.tight_layout()
plt.show()
```
